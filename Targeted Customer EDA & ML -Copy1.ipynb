```python
import pandas as pd
import numpy as np 
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.neighbors import KNeighborsClassifier
from sklearn.linear_model import LogisticRegression
from sklearn.model_selection import train_test_split, GridSearchCV,RandomizedSearchCV
from sklearn.metrics import recall_score,precision_score,f1_score,accuracy_score, confusion_matrix, classification_report
from sklearn.tree import DecisionTreeClassifier
from sklearn.ensemble import RandomForestClassifier
pd.set_option("display.max_columns", None)
```

```python
lr = pd.read_excel('/Users/chirag/Desktop/Office/HZ/YTD Lead Report.xlsx')
pr = pd.read_csv('/Users/chirag/Desktop/Office/HZ/YTD Payment Report.csv')
```

```python
pr
```

```python
l=lr[['Product Type','Partner name','Vehicle Make','Manufacturing Date','Previous Year Claim','Total Premium with GST','State']]

f=pr[['Product Type','Partner name','Vehicle Make','Manufacturing Date','Previous Year Claim','Total Premium with GST','State']]
```

```python
l.dropna(inplace=True)
```

```python
l['Sucess status'] = l['Sucess status']=0
```

```python
f['Sucess status'] = f['Sucess status']=1
```

```python
df=pd.concat([l,f])
```

```python
df
```

```python
df = df.sample(frac=1).reset_index(drop=True)
```

```python
df.dtypes
```

```python
df['Product Type'] = df['Product Type'].astype('category')
df['Product Type'] = df['Product Type'].cat.codes
df['Partner name'] = df['Partner name'].astype('category')
df['Partner name'] = df['Partner name'].cat.codes
df['Vehicle Make'] = df['Vehicle Make'].astype('category')
df['Vehicle Make'] = df['Vehicle Make'].cat.codes
df['Previous Year Claim'] = df['Previous Year Claim'].astype('category')
df['Previous Year Claim'] = df['Previous Year Claim'].cat.codes
df['State'] = df['State'].astype('category')
df['State'] = df['State'].cat.codes
```

```python
df
```

```python
df['Sucess status'].value_counts()
```

```python
X = df.iloc[:,:6]
y = df['Sucess status']
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X,y, test_size=0.3, random_state=0)
```

```python
from sklearn.linear_model import LogisticRegression
classification = LogisticRegression()
```

```python
classification.fit(X_train, y_train)
```

```python
y_pred = classification.predict(X_test)
```

```python
from sklearn import metrics
print(metrics.classification_report(y_test, y_pred))
```

```python
from sklearn.metrics import accuracy_score
accuracy_score(y_test, y_pred)
```
